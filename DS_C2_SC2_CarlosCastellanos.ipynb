Extracción de la información del archivo

```python
#Importamos la librerías
import pandas as pd

#Leemos el archivo csv
file =  pd.read_csv('/content/country_vaccinations.csv')

#Verificamos que se extrajo la información correctamente
file.head(5)
```

Estructura y tipo de datos de cada columna

```python
#Mostramos la estructura y tipo de datos
file.info()
```

```python
#Validamos que las columnas de fechas sean de tipo datetime64
file['date'] = pd.to_datetime(file['date'])

#Mostramos la estructura y tipo de datos con las columnas de fecha tipo datetime64
file.info()
```

Cantidad de vacunas aplicadas de cada compañia

```python
file[['vaccines', 'total_vaccinations']].groupby('vaccines').sum()
```

Cantidad de vacunas aplicada en todo el mundo

```python
file['total_vaccinations'].sum()
```

Promedio de vacunas aplicadas por país

```python
file[['country', 'total_vaccinations']].groupby('country').sum()
```

Cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
file[file['date'] == '2021-01-29']['total_vaccinations'].sum()
```

Dataframe nuevo llamado conDiferencias


```python
conDiferencias = file.copy()

conDiferencias['Diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']

conDiferencias.head(5)
```

Obtención del período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua. 

```python
# Obtenemos la fecha más antigua y la fecha más reciente
fecha_antigua = file['date'].min()
fecha_reciente = file['date'].max()

# Calculamos la diferencia entre ambas fechas
diferencia = fecha_reciente - fecha_antigua
print(diferencia)
```

Dataframe nuevo llamado conCantidad

```python
conCantidad = file.copy()

conCantidad['canVac'] = conCantidad['vaccines'].str.split('/').str.len()

conCantidad.head(5)
```

Dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020

```python
antes20 = file.copy()

antes20 = antes20[antes20['date'] < '2020-12-20']

antes20.head()
```

Dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer

```python
pfizer = file.copy()

pfizer = pfizer[pfizer['vaccines'].apply(lambda x:  'Pfizer' in x)]

pfizer.head(10)
```

Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel

```python
with pd.ExcelWriter('/content/resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias")  
    conCantidad.to_excel(writer, sheet_name="conCantidad")
    antes20.to_excel(writer, sheet_name="antes20")
    pfizer.to_excel(writer, sheet_name="pfizer")
```
